    <br>

    کمپین بازاریابی یا پویش بازاریابی، مجموعه فعالیت‌هایی است که آگاهی مشتریان را نسبت به یک محصول افزایش می‌دهد.
    <br>
     در مورد بازاریابی بانکی می‌توان این‌گونه تعریف کرد که متخصصان بازاریابی بانکی، با روش‌های متنوع سعی در افزایش آگاهی مشتریان درمورد خدمات بانکی دارند.
    در این مسئله نیز ما اطلاعات مختلفی از مشتریان یک بانک را در اختیار داریم. همچنین می‌دانیم نتیجه‌ی کمپین‌های قبلی مارکتینگ برای آن‌ها چگونه بوده است. بنابراین قصد داریم به روش‌های هوشمند پیش‌بینی کنیم کدام مشتر‌ی‌ها، برای سپرده‌‌ی بلندمدتی که تحت کمپین جدید است، اقدام می‌کنند. به عبارت بهتر، می‌خواهیم بدانیم نتیجه‌ی کمپین جدید، برای کدام مشتری موفقیت‌آمیز خواهد بود. 
    از آن جایی که هنوز با مدل‌سازی آشنا نشده‌اید صرفاً از شما می‌خواهیم در پیش‌پردازش داده‌ها به ما کمک کنید. کدهای مرتبط با مدل‌سازی داخل همین نوت‌بوک وجود دارند. شما فقط کافیست آن‌ها را اجرا کنید!

```python
import pandas as pd
from flaml import AutoML
from sklearn.metrics import f1_score
```

```python
df = pd.read_csv('../data/bank-additional-full.csv', na_values='unknown')
df.head(15)
```


<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    برای آن که ارزش کار خود را درک کنید، می‌توانید ابتدا که هیچ‌گونه پیش‌پردازشی روی داده انجام نشده است، یکبار با داده‌های اصلی مدل‌سازی کرده و عملکرد مدل را مشاهده کنید. با اجرای سلول زیر یک مدل روی داده‌ها آموزش می‌بیند و نتایجی تولید می‌کند. توجه کنید که اجرای این سلول و فرآیند مدل‌سازی ممکن است چند دقیقه‌ای طول بکشد.
    <br>
    اگر اجرای این سلول با شکست مواجه شد، می‌تواند به یکی از دلایل زیر باشد.
</font>
</p>


<div dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    <ul>
        <li>
            کتابخانه‌ی Fast Auto ML را نصب نکرده باشید.
        </li>
        <li>
            در بارگیری مجموعه‌داده دچار مشکل شده باشید.
        </li>
    </ul>
</font>
</div>

```python
target_variable = 'y'
df_train = df.sample(frac=.7, random_state=313)

model = AutoML(task='classification', time_budget=120, verbose=0)
model.fit(df_train.drop(target_variable, axis=1), df_train[target_variable])

df_test = df.drop(df_train.index)
y_pred = model.predict(df_test.drop(target_variable, axis=1))

f1score = f1_score(df_test[target_variable], y_pred, pos_label='yes')*100
print(f'performance of model is {f1score} %')
```

<h2 align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
قسمت اول
</font>
</h2>

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
همان‌طور که قبلا گفته شد، در ستون‌های دسته‌ای این دیتافریم، مقادیر گم‌شده وجود دارد. به عنوان قسمت اول از این سوال، از شما می‌خواهیم:
</font>
</p>
<ol dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
    <li>اسم همه ستون‌های دسته‌ای دیتافریم را درون لیستی به اسم <code>categorical_cols</code> بریزید. متغیر هدف نباید در این لیست قرار گرفته باشد.</li>
    <li>سپس به ازای همه ستون‌هایی که اسم‌شان در <code>categorical_cols</code> وجود دارد، مقادیر گم‌شده هر ستون را با پر تکرارترین مقدار آن ستون جایگزین کنید. دیتافریمی که هیچ مقدار گم‌شده‌ای ندارد را درون خود دیتافریم <code>df</code> ذخیره کنید. بنابراین بعد از حل قسمت اول، دیتافریم <code>df</code> هیچ مقدار گم‌شده‌ای نباید داشته باشد.</li>
</ol>


```python
df.info()
```

```python
# TO-DO: fill `categorical_cols` and fill na values in categorical columns of `df` dataframe.
categorical_cols = list(df.columns)[1:10] + list(df.columns)[-2: -1]
for col in categorical_cols:
    df[col].fillna(value= df[col].mode()[0], inplace= True)
    
categorical_cols
```

<h2 align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
قسمت دوم
</font>
</h2>

<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
در این قسمت از شما انتظار داریم ۴ لیست آماده کنید. در هر لیست، ستون‌های دیتافریم <code>df</code> به تفکیک نوع‌شان طبقه‌بندی شده‌اند.

<ul dir=rtl>
    <li>در لیست <code>continuous_cols</code> اسامی ستون‌هایی را بنویسید که مقدار پیوسته دارند.</li>
    <li>در لیست <code>binary_cols</code> اسامی ستون‌هایی را بنویسید که مقدار گسسته دودویی دارند.</li>
    <li>در لیست <code>nominal_cols</code> اسامی ستون‌هایی را بنویسید که مقدار اسمی دارند.</li>
    <li>در لیست <code>ordinal_cols</code> اسامی ستون‌هایی را بنویسید که مقدار ترتیبی دارند.</li>
</ul>
</font>
</p>

<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
توجه کنید که نوع (<code>dtype</code>) تمام متغیر‌هایی که در بالا معرفی شدند، باید حتما لیست باشد.
</font>
</p>

```python

continuous_cols = ["age", "duration", "campaign", "pdays", "previous"]
binary_cols = ["default", "housing", 'loan',  'is_telephone_contact']
nominal_cols = ['job', 'marital']
ordinal_cols = ['education', 'day_of_week', 'month', 'poutcome']
df.head()
```

<h2 align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
قسمت سوم
</font>
</h2>

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
در قسمت سوم از شما می‌خواهیم دیتافریمی داشته باشید که فقط شامل ستون‌های دودویی دیتافریم <code>df</code> باشد. سپس همه مقادیر دیتافریم جدید را به عدد تبدیل کنید.
<br>
با استفاده از <code>binary_cols</code> می‌توانید دیتافریم <code>binary_df</code>  را بسازید.
در حقیقت دیتافریم <code>binary_df</code> بخشی از دیتافریم <code>df</code> است که فقط شامل ستون‌های دودویی آن باشد. یعنی هیچ ستون دیگری با مقادیر پیوسته، اسمی یا ترتیبی درون دیتافریم <code>binary_df</code> وجود ندارد. توجه کنید که تمام مقادیر این دیتافریم باید عدد باشند (به‌جای <code>yes</code> عدد <code>1</code> و به جای <code>no</code> عدد <code>0</code> قرار دهید).
</font>
</p>

```python
# TO-DO: create `binary_df` and replace all `object` values with numeric ones
binary_df = pd.DataFrame()
for col in binary_cols:
    binary_df[col] = df[col].apply(lambda x: 1 if x == 'yes' else 0)
    
binary_df
```
